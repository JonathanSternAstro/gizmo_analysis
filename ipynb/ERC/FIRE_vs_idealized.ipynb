```python
import numpy as np, pylab as pl
from matplotlib import ticker
```

```python
Xs = Ys = np.linspace(-50,50,500)
FIRE = np.load('stacked_fire_data.npy')
ideal = np.load('idealized_data.npy')
```

```python
pl.plot(Xs,FIRE[250,:],c='k')
pl.plot(Xs,ideal[250,:])
```

```python
pl.figure(figsize=(7,3)); pl.subplots_adjust(wspace=0.02,left=0.01,right=0.99)
axs = pl.subplot(121), pl.subplot(122)
for iPanel in range(2):
    ax = axs[iPanel]; pl.sca(ax)
    vals = (ideal,FIRE)[iPanel]
    title = ('idealized simulation','FIRE stack')[iPanel]
    pl.contour(Xs,Ys,vals,[-50,-25,0,25,50],colors='.5')
    mesh = pl.pcolormesh(Xs,Ys,vals,vmin=-200,vmax=200,cmap='seismic_r')    
    pl.title(title)
    pl.xlim(-40,40); pl.ylim(-40,40)
    [axis.set_major_locator(ticker.NullLocator()) for axis in (ax.xaxis,ax.yaxis)]
pl.colorbar(ax=axs)
    
```
